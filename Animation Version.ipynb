```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

```python
nx, ny = 50, 50
T = np.zeros((nx, ny))

# Heat source (center)
T[25, 25] = 100
```

```python
alpha = 0.1
```

```python
fig, ax = plt.subplots()
heatmap = ax.imshow(T, cmap='hot', vmin=0, vmax=100)
plt.colorbar(heatmap)
ax.set_title("2D Heat Diffusion Animation")
```

```python
def update(frame):
    global T
    T_new = T.copy()
    
    for i in range(1, nx-1):
        for j in range(1, ny-1):
            T_new[i, j] = T[i, j] + alpha * (
                T[i+1, j] + T[i-1, j] +
                T[i, j+1] + T[i, j-1] -
                4*T[i, j]
            )
    
    T = T_new
    heatmap.set_array(T)
    return [heatmap]
```

```python
from IPython.display import HTML

ani = FuncAnimation(fig, update, frames=100, interval=100)

HTML(ani.to_jshtml())
```
