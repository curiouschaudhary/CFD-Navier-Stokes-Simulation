```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Grid size
nx, ny = 50, 50

# Temperature grid (all zeros initially)
T = np.zeros((nx, ny))
```

```python
# Set center point hot
T[25, 25] = 100
```

```python
alpha = 0.1   # diffusion coefficient
timesteps = 100
```

```python
for t in range(timesteps):
    T_new = T.copy()
    
    for i in range(1, nx-1):
        for j in range(1, ny-1):
            T_new[i, j] = T[i, j] + alpha * (
                T[i+1, j] + T[i-1, j] +
                T[i, j+1] + T[i, j-1] -
                4*T[i, j]
            )
    
    T = T_new
```

```python
plt.imshow(T, cmap='hot')
plt.colorbar()
plt.title("2D Heat Diffusion Simulation")
plt.show()
```
